__9. Kalle ska bygga en ML-modell och delar upp sin data i ”Träning”, ”Validering” och ”Test”, vad används respektive del för?__
Träning används för att träna modellen genom fit. Modellen lär sig mönster från datan och sätter parametrar. Här kan man träna många olika modeller men det är först mot valideringsdata man ser om modellen presterar bra. 
Validering används för att testa prediktioner som modellen har tränat fram genom fit på träningsdatan, hur bra den gissar på data den inte sett än. Här jämför vi modeller mot varandra och tar fram bra hyperparametrar med ex GridSearchCV. När prediktionerna är tillräckligt bra väljer man modell att gå vidare med. 
Test används i slutsteget en gång för att se om den valda modellen predikterar bra mot osedd data. Detta ska ses som en slutkontroll på hur bra modellen generaliserar. Om resultatet är dåligt får man börja om processen från början och förbättra modellen eller byta modell. Viktigt att när man gör om processen inte använda testdata för att ta beslut.

__14. Vad är en ”Confusion Matrix”?__
En matrix som visuellt visar hur bra en modell presterar genom att jämföra sanna (facit) och predikterade värden, hur väl modellens prediktioner fördelar sig jämt med de sanna etiketterna. 
Man kommer att i varje ruta i matrisen se hur många gånger modellen predikterat värden, både korrekt och felaktigt.

Binert klassificeringsprblem: 
True Positive TP (facit: positivt, modellen predikterar positivt = rätt)
True Negative TN (facit: negativt, modellen predikterar negativt = rätt)
False Negative FN (facit: positivt, modellen predikterar negativt = fel)
False Positive FP (facit: negativt, modellen predikterar positivt = fel)

TP | FN
- - - -
FP | TN


__15. Vad är K-means modellen för något? Ge ett exempel på vad det kan tillämpas på.__
Den vanligaste (icke vägledda) modellen inom klustring (när datapunkter delas in i olika grupper/kluster där datapunkter som liknar varandra ska hamna i samma kluster). Modellen är ofta iterativ och går ut på att varje datapunkt tilldelas det kluster vars centroid är närmast. Efter varje iteration uppdateras centroidernas position/koordinater så att dom hamnar i mitten av sina tilldelade datapunkter. 
Snabb och skalbar, kan användas på stora dataset. Den presterar bäst när storlek på kluster är ungefär lika stora, när klustren har en sfärisk form, klustren har liknande spridning. 

Används ex i Kundsegmentering, för att dela in olika kunder med samma attribut i samma kluster (ex köpbeteende el demografi). 



__16. Förklara (gärna med ett exempel): Ordinal encoding, one-hot encoding, dummy variable encoding.__
Ordinal encoding: när vi omvandlar kategorisk data till numerisk data där värdena får en inbördes rangordning. Ex låg, medel, hög kanske ska viktas olika och genom ordinal encoding kan vi få dom att rangordnas till ex Låg=1, Medel=2, Hög=3. 
One-hot encoding: när vi omvandlar kategorisk data till numerisk data men där värdena inte har någon inbördes rangordning. Ex vi har färger på en bil vit, svart, röd, blå. Det skapas en kolumn per kategori (färg i detta exemplet) och per kategori visa 1 eller 0, 1 om bilen har kategorifärgen färgen och 0 om den inte har den färgen. 
Dummy variable encoding: när vi omvandlar kategorisk data till numerisk data men där värdena inte har någon inbördes rangordning. Skillnaden från one-hot encoding är att vi skapar en kolumn per kategori minus 1 (k-1), den kategori som vi tar bort blir en referenskategori. Viktigast i Linjära modeller som Linear Regression oh Logistisk Regression. Används för att undvika redundans och göra linjära modellen stabilare och lättare att tolka.

__18. Vad är skillnaden mellan parametrar och hyperparametrar i en maskininlärningsmodell? Ge ett exempel på varje och förklara varför de inte kan optimeras på samma sätt.__
Parametrar: "Vad vi lär oss". Lärs in automatiskt av modellen under .fit. De bestäms av träningsalgoritmen för att minimera felet. Ex om jag kör Standardscaler() på min data så lär den sig medelvärde per feature, standardavvikelse/skalning per feature. Det är parametrarna som modellen lär sig automatiskt från datan. 
Hyperparametrar: "Hur vi lär oss". ställs in manuellt av den som tränar modellen innan .fit. Dessa används för att för att styra träningsprocessen/modellens beteende och komplexitet. Ex för SVM kan man sätta C- och gamma-värden. Använder man sig av GridSearchCV kan man testa olika hyperparametrar och välja dom som ger bäst resultat, bäst prediktioner.

Parametrar optimeras direkt inne i träningen automatiskt under .fit. 
Hyperparametrar är någoting du manuellt väljer utifrån hur bra modellen generaliserar. Iterativ process där du testar olika parametrar för att hitta fram till dom som passar din modell bäst. 



__19. Förklara skillnaden mellan overfitting och underfitting i en maskininlärningsmodell. Beskriv även hur man kan upptäcka respektive åtgärda dem.__
Hur kan man upptäcka: 
Overfitting: att modellen presterar för bra på din träningsdata, att den lär sig sin träningsdata för bra. Att den hittar alla värden i din träningsdata men när du testar den mot valideringsdata presterar den dåligt. Hög train score men låg validation score. 
Underfitting: att modellen presterar för dåligt på din data. Att det är en för enkel modell som är dålig på att hitta mönster. Dålig på både träningsdata och valideringsdata. Låg train score och låg validation score. 

Hur kan man åtgärda: 
Overfitting -> minskar modellens komplexitet, tuning av hyperparametrar. 
Underfitting -> ökar modellens komplexitet, lägga till bättre features, byt till en kraftfullare modell

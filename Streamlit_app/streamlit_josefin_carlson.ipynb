```python
# Importerar bibliotek
import joblib
import numpy as np
import sklearn
import sys

from sklearn.datasets import fetch_openml
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.svm import SVC


# Hämtar in data för MNIST och sparar som X och y
mnist = fetch_openml('mnist_784', version = 1, cache = True, as_frame = False)

X = mnist["data"].astype("float32")
y = mnist["target"].astype(int)
```

```python
BEST_PCA = 75
BEST_C = 14
BEST_GAMMA = "scale"

final_model_svc = Pipeline([
    ("scaler", StandardScaler()),
    ("pca", PCA(n_components=BEST_PCA, random_state=42, svd_solver="randomized")),
    ("model", SVC(kernel="rbf", C=BEST_C, gamma=BEST_GAMMA))
])

#Tränar modellen på hela dataset
final_model_svc.fit(X, y)

#Sparar modellen
joblib.dump(final_model_svc, "svc_mnist.joblib")

meta = {
    "sklearn_version": sklearn.__version__,
    "numpy_version": np.__version__,
    "python_version": sys.version,
    "input_shape": (784,),
    "expected_input": "28x28 grayscale, values 0-255, float32, flattened to (1,784)",
    "pca_components": BEST_PCA,
    "C": BEST_C,
    "gamma": BEST_GAMMA
}

joblib.dump(meta, "svc_mnist_meta.joblib") 

print("Sparat: svc_mnist.joblib och svc_mnist_meta.joblib")
```
